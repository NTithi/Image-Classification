```python
import cv2
from google.colab.patches import cv2_imshow

# Read the image from the specified path
img = cv2.imread("/content/drive/MyDrive/Computer Vision work/img1.jpg")

# Check if the image was read successfully
if img is None:
    print("Error: Image not found or unable to read.")
else:
    # Resize the original image to a smaller size, e.g., (400, 300)
    img_resized = cv2.resize(img, (400, 300))

    # Convert to grayscale and resize
    ImgGray = cv2.cvtColor(img_resized, cv2.COLOR_BGR2GRAY)

    # Apply Gaussian blur and resize
    ImgBlur = cv2.GaussianBlur(ImgGray, (7,7), 0)

    # Display the resized images
    print("Gray Image (Resized):")
    cv2_imshow(ImgGray)  # Show resized grayscale image

    print("Original Image (Resized):")
    cv2_imshow(img_resized)  # Show resized original image

    print("Blurred Image (Resized):")
    cv2_imshow(ImgBlur)  # Show resized blurred image
```

```python
import cv2
from google.colab.patches import cv2_imshow

# Read the image from the specified path
img = cv2.imread("/content/drive/MyDrive/Computer Vision work/img1.jpg")

# Check if the image was read successfully
if img is None:
    print("Error: Image not found or unable to read.")
else:
    # Resize the original image to a smaller size, e.g., (400, 300)
    img_resized = cv2.resize(img, (400, 300))

    # Convert to grayscale and resize
    ImgGray = cv2.cvtColor(img_resized, cv2.COLOR_BGR2GRAY)

    # Apply Gaussian blur and resize
    ImgBlur = cv2.GaussianBlur(ImgGray, (7,7), 0)

    # Convert grayscale images to 3-channel to concatenate
    ImgGray_3ch = cv2.cvtColor(ImgGray, cv2.COLOR_GRAY2BGR)
    ImgBlur_3ch = cv2.cvtColor(ImgBlur, cv2.COLOR_GRAY2BGR)

    # Concatenate the images horizontally
    combined = cv2.hconcat([ImgGray_3ch, img_resized, ImgBlur_3ch])

    # Display the combined image
    cv2_imshow(combined)
```
